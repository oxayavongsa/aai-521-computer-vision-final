```python
import os
import random
from easyocr import Reader

# Define the folder containing the images
extract_dir = '/content/extracted_images/output_images/unzipped_dataset/train_val_images/train_images'  # Replace with your destination folder

# Step 1: Initialize EasyOCR reader
reader = Reader(['en'], gpu=False)  # Set gpu=True if you have a compatible GPU

# Step 2: Define function to process a subset of images
def process_images(image_dir, reader, percentage=10):
    # List all image files in the directory
    image_files = [f for f in os.listdir(image_dir) if f.endswith(('.png', '.jpg', '.jpeg'))]

    # Shuffle and select 10% of the images
    sample_size = max(1, len(image_files) * percentage // 100)  # Ensure at least one image is processed
    sampled_files = random.sample(image_files, sample_size)

    extracted_texts = {}

    for image_file in sampled_files:
        image_path = os.path.join(image_dir, image_file)
        results = reader.readtext(image_path)

        # Concatenate detected text
        extracted_text = " ".join([res[1] for res in results])
        extracted_texts[image_file] = extracted_text

        print(f"Processed {image_file}: {extracted_text}")

    return extracted_texts

# Step 3: Process 10% of the images
extracted_texts = process_images(extract_dir, reader, percentage=10)

# Step 4: Save results to a file
results_file = 'extracted_texts_sampled.txt'
with open(results_file, 'w') as f:
    for image_name, text in extracted_texts.items():
        f.write(f"{image_name}: {text}\n")

print(f"Extracted texts saved to {results_file}")
```

```python
!pip install python-Levenshtein
import pandas as pd
from Levenshtein import distance as levenshtein_distance

# Step 1: Load ground truth from annotated.csv
csv_file = '/content/extracted_images/output_images/unzipped_dataset/annot.csv'  # Path to your CSV file
annotated_data = pd.read_csv(csv_file)

# Ensure all 'utf8_string' values are strings and handle missing values
annotated_data['utf8_string'] = annotated_data['utf8_string'].fillna("").astype(str)

# Create a dictionary for ground truth
ground_truth = dict(zip(annotated_data['id'], annotated_data['utf8_string']))

# Step 2: Load extracted texts from TXT file
extracted_texts = {}
txt_file = 'extracted_texts_sampled.txt'  # Replace with the correct path

with open(txt_file, 'r') as file:
    for line in file:
        line = line.strip()
        if ": " in line:  # Ensure the line contains the delimiter
            image_name, text = line.split(": ", 1)
            extracted_texts[image_name] = str(text)  # Force string conversion
        else:
            print(f"Skipping malformed line: {line}")

# Step 3: Evaluate OCR results
exact_matches = 0
total_distance = 0
word_accuracies = []

for image_id, true_text in ground_truth.items():
    # Convert true_text to string to handle any non-string values
    true_text = str(true_text)

    # Get extracted text and ensure it's a string
    extracted_text = extracted_texts.get(image_id, "")
    extracted_text = str(extracted_text)

    # Exact match evaluation
    if true_text.strip() == extracted_text.strip():
        exact_matches += 1

    # Levenshtein distance
    total_distance += levenshtein_distance(true_text, extracted_text)

    # Word-level accuracy
    true_words = set(true_text.split())
    extracted_words = set(extracted_text.split())
    if true_words:
        word_accuracy = len(true_words & extracted_words) / len(true_words)
        word_accuracies.append(word_accuracy)
    else:
        word_accuracies.append(0)

# Step 4: Calculate metrics
exact_match_accuracy = exact_matches / len(ground_truth) * 100
average_levenshtein_distance = total_distance / len(ground_truth)
average_word_accuracy = sum(word_accuracies) / len(word_accuracies) * 100

# Step 5: Print results
print(f"Exact Match Accuracy: {exact_match_accuracy:.2f}%")
print(f"Average Levenshtein Distance: {average_levenshtein_distance:.2f}")
print(f"Average Word-Level Accuracy: {average_word_accuracy:.2f}%")
```
